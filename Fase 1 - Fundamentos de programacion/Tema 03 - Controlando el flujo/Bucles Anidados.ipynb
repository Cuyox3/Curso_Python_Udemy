```python
lista = [ 
    "Esto es un texto",             #cadena 
    (1,5,10,15,20,25),              #tupla
    ["Azul","Verde","Amarillo"]    #lista
]
```

```python
for coleccion in lista: 
    for elemento in coleccion:
        print (coleccion,"->",elemento)
```

```python
print ("Introuzca un numero entre el 1 y el 9 ")
multiplos = []
numero = int (input())
while 9>numero:
    for i in range (1,101):
        if numero%i == 0 : 
            multiplos.append(i)
    print(multiplos)
```

```python
for indice_coleccion, coleccion in enumerate (lista):
    for indice_elemento, elemento in enumerate (coleccion): 
        print (coleccion,"->",elemento)
```

```python
tabla = [
    [0,0,0],
    [1,1,1],
    [2,2,2]
]
for fila in tabla: 
    for columna in fila: 
        print (columna, end="")# end se puede poner para agregar algo al final
    print ()# imprime un salto de linea para mostrar de nuevo una tabla 
```

```python
tabla = [
    [0,0,0],
    [1,1,1],
    [2,2,2]
]
for i, fila  in enumerate(tabla): #Se agrag la varibale de enumeracion parael enumerador  
    for j, columna in enumerate(fila): 
        print (tabla[i][j], end=" " )
    print ("")
```
