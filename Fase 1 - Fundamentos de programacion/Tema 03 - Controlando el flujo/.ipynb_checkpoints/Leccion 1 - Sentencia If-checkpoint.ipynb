```python
nota =float (input("Introduce una nota:"))
if nota >= 9:
    print ("Sobresalientne")
elif nota >= 7 and nota < 9:
    print ("Notable")
elif nota >= 6:
    print ("Bien")
elif nota >= 5 :
    print ("Suficiente")
else:
    print ("Insuficiente")
```
