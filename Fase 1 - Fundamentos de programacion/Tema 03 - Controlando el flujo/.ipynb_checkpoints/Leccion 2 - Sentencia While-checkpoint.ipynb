```python
print ("Es bienvenido al bucle interactivo")
while (True): 
    print (""""Que quieres hacer? Escriba na opcion
    1)Saludar 
    2)Sumar dos datos 
    3) Salir """)
    opcion = input()
    if opcion == '1':
        print ("Hola espero que estes pasando un buen dia")
    elif opcion == '2': 
        n1 =float (input("Introduce el primer numero "))
        n2 =float (input("Introduce el segundo numero "))
        print ("El resultado de la suma es:", n1+n2)
    elif opcion == '3':
        print ("Hasta pronto! Ha saido un placer!")
        
    else: 
        print ("Comando desconocido")
```
